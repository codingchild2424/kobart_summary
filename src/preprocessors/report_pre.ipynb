```python
report_path1 = "/workspace/home/uglee/Projects/kobart_summary/datasets/summary/report1"
report_path2 = "/workspace/home/uglee/Projects/kobart_summary/datasets/summary/report2"

import os
import json
from tqdm import tqdm
import re

pre_data_path = "/workspace/home/uglee/Projects/kobart_summary/datasets/preprocessed"

for data_path in [report_path1, report_path2]:
    for text_type in os.listdir(data_path):

        try:
            # short
            short_sent_path = os.path.join(data_path, text_type, "2~3sent")

            for json_file_name in os.listdir(short_sent_path):
                json_file_path = os.path.join(short_sent_path, json_file_name)

                # open the json file
                with open(json_file_path, "r") as f:
                    # read the json file
                    json_data = json.load(f)
                    
                    title = json_data['Meta(Acqusition)']['doc_name']
                    title = title.replace("\t", "")
                    title = title.replace("\n", " ")
                    # using re title preprocess special character
                    title = re.sub('[^a-zA-Z0-9ㄱ-힗]', '', title)
                    # remove space
                    title = title.replace(" ", "")
                    # title lenght is over 20, cut the title
                    if len(title) > 20:
                        title = title[:20]
                        
                    # summary_entire
                    content1 = json_data['Meta(Refine)']["passage"]
                    summary1 = json_data["Annotation"]["summary1"]

                    # preprocess
                    content1 = content1.replace("\t", "")
                    content1 = content1.replace("\n", " ")

                    summary1 = summary1.replace("\t", "")
                    summary1 = summary1.replace("\n", " ")

                    content_summary1 = content1 + "\t" + summary1

                    tsv_file_name1 = title + "1" + ".tsv"

                    tsv_file_path1 = os.path.join(pre_data_path, tsv_file_name1)

                    # save the tsv file
                    with open(tsv_file_path1, "w") as t1:
                        t1.write(content_summary1)


                    # summary_section
                    content2 = json_data['Meta(Refine)']["passage"]
                    summary2 = json_data["Annotation"]["summary2"]

                    # preprocess
                    content2 = content2.replace("\t", "")
                    content2 = content2.replace("\n", " ")

                    summary2 = summary2.replace("\t", "")
                    summary2 = summary2.replace("\n", " ")

                    content_summary2 = content2 + "\t" + summary2

                    tsv_file_name2 = title + "2" + ".tsv"

                    tsv_file_path2 = os.path.join(pre_data_path, tsv_file_name2)

                    # save the tsv file
                    with open(tsv_file_path2, "w") as t2:
                        t2.write(content_summary2)


            # long
            long_sent_path = os.path.join(data_path, text_type, "20per")

            for json_file_name in os.listdir(long_sent_path):
                json_file_path = os.path.join(long_sent_path, json_file_name)

                # open the json file
                with open(json_file_path, "r") as f:
                    # read the json file
                    json_data = json.load(f)
                    
                    title = json_data['Meta(Acqusition)']['doc_name']
                    title = title.replace("\t", "")
                    title = title.replace("\n", " ")
                    # using re title preprocess special character
                    title = re.sub('[^a-zA-Z0-9ㄱ-힗]', '', title)
                    # remove space
                    title = title.replace(" ", "")
                    # title lenght is over 20, cut the title
                    if len(title) > 20:
                        title = title[:20]
                        
                    # summary_entire
                    content1 = json_data['Meta(Refine)']["passage"]
                    summary1 = json_data["Annotation"]["summary1"]

                    # preprocess
                    content1 = content1.replace("\t", "")
                    content1 = content1.replace("\n", " ")

                    summary1 = summary1.replace("\t", "")
                    summary1 = summary1.replace("\n", " ")

                    content_summary1 = content1 + "\t" + summary1

                    tsv_file_name1 = title + "1" + ".tsv"

                    tsv_file_path1 = os.path.join(pre_data_path, tsv_file_name1)

                    # save the tsv file
                    with open(tsv_file_path1, "w") as t1:
                        t1.write(content_summary1)


                    # summary_section
                    content2 = json_data['Meta(Refine)']["passage"]
                    summary2 = json_data["Annotation"]["summary3"]

                    # preprocess
                    content2 = content2.replace("\t", "")
                    content2 = content2.replace("\n", " ")

                    summary2 = summary2.replace("\t", "")
                    summary2 = summary2.replace("\n", " ")

                    content_summary2 = content2 + "\t" + summary2

                    tsv_file_name2 = title + "2" + ".tsv"

                    tsv_file_path2 = os.path.join(pre_data_path, tsv_file_name2)

                    # save the tsv file
                    with open(tsv_file_path2, "w") as t2:
                        t2.write(content_summary2)

        except:
            pass
```
