```python
patent_path1 = "/workspace/home/uglee/Projects/kobart_summary/datasets/summary/patent1"
patent_path2 = "/workspace/home/uglee/Projects/kobart_summary/datasets/summary/patent2"
patent_path3 = "/workspace/home/uglee/Projects/kobart_summary/datasets/summary/patent3"

import os
import json
from tqdm import tqdm
import re

pre_data_path = "/workspace/home/uglee/Projects/kobart_summary/datasets/preprocessed"

for data_path in [patent_path1, patent_path2, patent_path3]:
    for json_file_name in os.listdir(data_path):

        try:
            # folder path
            json_file_path = os.path.join(data_path, json_file_name)

            # open the json file
            with open(json_file_path, "r") as f:
                # read the json file
                data = json.load(f)

                for json_data in data["data"]:
                    
                    title = json_data['title']
                    title = title.replace("\t", "")
                    title = title.replace("\n", " ")
                    # using re title preprocess special character
                    title = re.sub('[^a-zA-Z0-9ㄱ-힗]', '', title)
                    # remove space
                    title = title.replace(" ", "")
                    # title lenght is over 20, cut the title
                    if len(title) > 20:
                        title = title[:20]
                        
                    # summary_entire
                    content = json_data['summary_section'][0]["orginal_text"]
                    summary = json_data["summary_section"][0]["summary_text"]

                    # preprocess
                    content = content.replace("\t", "")
                    content = content.replace("\n", " ")

                    summary = summary.replace("\t", "")
                    summary = summary.replace("\n", " ")

                    content_summary = content + "\t" + summary

                    tsv_file_name = title + "1" + ".tsv"

                    tsv_file_path = os.path.join(pre_data_path, tsv_file_name)

                    # save the tsv file
                    with open(tsv_file_path, "w") as t:
                        t.write(content_summary)
        except:
            pass
                    
```
